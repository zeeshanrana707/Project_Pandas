# Element Slicing using loc and iloc

```python
# Single element slicing loc
df.loc[0, 'Country']
```

```python
# Single element slicing using iloc
df.iloc[2, 3]
```

```python
# Multiple element slicing using loc
# Select Multiple rows and columns
df.loc[0:3, 'Country': 'Complaints_20']
```

# Data Cleaning 

```python
#Finding Null Values
# Check for Null Values
df.isnull()
```

```python
#  Count Null Values
#  Count Null Values in each column
df.isnull().sum()
```

```python
# Count total null values in dataframe
df.isnull().sum().sum()
```

```python
# finding rows with null values
# Identify rows containing any null values
df.isnull().any(axis=1)
```

```python
# Identify rows where all columns are null
df.isnull().all(axis=1)
```

```python
# Find Columns with Null Values
df.isnull().any()
```

```python
# Fill Null Values with Column Mean
# For a single element
df["Complaints_20"].fillna(df["Complaints_20"].mean())
```

```python
# Group by relation
df.groupby('Country')['Complaints_20'].sum()
```
